```python
x8=np.loadtxt('8x8.dat')
T8 = x8[:,0]
E8 = x8[:,1]/64
mag8 = x8[:,3]/64
varE8 = x8[:,2]/64/64 - (x8[:,1]/64)**2
stde8 = varE8**0.5
stee8=stde8/((100000)**0.5)
varm8 = x8[:,4]/64/64 - (x8[:,3]/64)**2
stdm8 = varm8**0.5
stem8=stdm8/((100000)**0.5)
```

```python
plt.plot(T8, E8, 'x')
plt.errorbar(T8,E8,yerr=stee8)
xlabel("Temperature")
ylabel("Energy per spin")
```

```python
plt.plot(T8, mag8, 'x')
plt.errorbar(T8,mag8,yerr=stem8)
xlabel("Temperature")
ylabel("Magnetisation per spin")
```
