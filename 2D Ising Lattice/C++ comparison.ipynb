```python
data = np.loadtxt("2x2C.dat")#2x2 C++ generated file
T2 = data[:,0] #get the first column
C2 = data[:,5] # get the fifth column

#first we fit the polynomial to the data
fit = np.polyfit(T2, C2, 9)

#now we generate interpolated values of the fitted polynomial over the range of our function
T_min2 = np.min(T2)
T_max2 = np.max(T2)
T_range2 = np.linspace(T_min2, T_max2, 1000) #generate 1000 evenly spaced points between T_min and T_max
fitted_C_values2 = np.polyval(fit, T_range2) # use the fit object to generate the corresponding values of C
plt.plot(T2, C2, 'x', label='Data')
plt.plot(T_range2, fitted_C_values2, label='Fitted polynomial')
xlabel('Temperature')
ylabel('Heat Capacity')
plt.legend()
```
