#### Plot the eigenstates of the excited state surface that are being excited

The potential curves used in this problem are non-trivial, and the eigenstates of the excited electronic state are not known *a priori*. That is why a traditional approach to calculating the spectrum from Frank-Condon factors from the eigenstates of ground and excited states is less convenient than the time-dependent approach used here.

However, once the dynamics of the system is known, the eigenstates of excited state surfaces can be recovered: if the time-dependent Schrödinger equation is solved, it is little work to find the solutions of the time-independent Schrödinger equation. (D. J. Tannor, *Introduction to Quantum Mechanics: a time-dependent perspective*, University Science Books, 2007, section 6.3.) To see this we first expand the nuclear wavefunctions of the excited state curve at time t<sub>0</sub> &chi;<sup>1</sup>(x,t<sub>0</sub>), as a function of the, as of now unknown but existing, eigenstates of that curve &Lambda;<sub>v</sub><sup>1</sup>(x):

$$\chi^1(x,t_0)=\sum_i c_i \Lambda_i^1(x).$$

If we apply the excited state time-evolution operator to both sides of this expression (remember &hbar;=1), we get:

$$e^{-i \hat{H}^1 (t-t_0)} \chi^1(x,t_0)=e^{-i \hat{H}^1 (t-t_0)}\sum_i c_i \Lambda_i^1(x) \Leftrightarrow \chi^1(x,t)=\sum_i c_i e^{-i E_i^1 t}\Lambda_i^1(x).$$

By multiplying both sides of the equality by $e^{i E_v^1 t}$ where $E_v^1$ is an eigenvalue of a specific eigenstate, and integrate over time:

$$\int_{-\infty}^{\infty}e^{i E_v^1 t} \chi^1(x,t) dt= \sum_i c_i \int_{-\infty}^{\infty}e^{i (E_v^1-E_i^1) t} \Lambda_i^1(x) dt = 2 \pi \sum_i c_i \delta(E_v^1-E_i^1) \Lambda_i^1(x) dt = 2 \pi c_v \Lambda_v^1(x).$$

The left-hand side of the equation can be calculated from the dynamics, the right-had side is our unknown.

Plot, and comment on the nature of, the eigenfunctions that are being populated in the vibronic spectrum of this system.

### Technical tips

* To make sure you have defined the functions for the potential correctly, start by plotting them to reproduce the figure above.
* In order to obtain valid results it is critical to set up an adequate grid of space points. The grid should be fine enough to describe high momentum components of the wavefunction. You should set the limits of the grid such that the wavefunction remains equal to zero at the limits of the grid throughout all the simulation to avoid artifacts. Check that this is the case before doing any further calculations.
* The time-step should be small enough to propagate the wavefunction correctly. But a too small time-step may require too many steps to be taken, and depending on how fine is your space grid this may make the computation unfeasible on your computer. Make tests with short runs, and simulate for longer times when needed.
