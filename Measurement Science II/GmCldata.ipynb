```python
import numpy as np
```

```python
data=loadtxt("GmCl1_conc.dat", skiprows=2)
```

```python
plot(data[:,0], data[:,1], linestyle='',marker='o', markersize=5)
ylabel("Intensity")
xlabel("GmCl Concentration /mM")
show()
```

```python
data[:,1]
```

```python
Fu=1-(data[1:6,1]-data[6,1])/(data[0,1]-data[6,1])
plot(data[1:6,0], Fu,linestyle='',marker='o', markersize=5)
ylabel("Fu")
xlabel("GmCl Concentration /mM")
show()
```

```python
ΔG=-8.31*298*np.log(Fu/(1-Fu))
ΔG
```

```python
approx=polyfit(data[1:6,0], ΔG,1,cov=True )
params=approx[0]
m=params[0]
c=params[1]
approx
```

```python
err_arr=[]
for i in range(0,5):
    percent=data[1:6,3][i]/data[1:6,1][i]
    err_arr=err_arr+[percent]
```

```python
error=err_arr*ΔG
error
```

```python
errorbar(data[1:6,0],ΔG,yerr=error, linestyle='', marker='x', markersize=5)
plot(data[1:6,0], m*data[1:6,0]+c)
ylabel("ΔG")
xlabel("GmCl Concentration /mM")
show()
```

```python
spontconc=-c/m
spontconc
```
